```python
# --- SIMULADOR DE IA SIMBÓLICA: SISTEMA EXPERTO MINERO ---
# Este programa utiliza lógica de árbol de decisión para predecir riesgos y éxitos.

print("--- BIENVENIDO AL SISTEMA DE INTELIGENCIA ARTIFICIAL MINERA ---")
print("Por favor, ingrese los datos de la excavación para que la IA realice una predicción.")

# Entrada de datos interactiva (Casteo a entero para valores numéricos)
mineral = input("\n¿Qué mineral busca el minero? (Oro / Carbón / Hierro): ")
profundidad = int(input("¿A qué profundidad (metros) se encuentra la veta?: "))
nivel_gas = int(input("Ingrese el nivel de gas detectado (0 a 100): "))

# Lógica de la IA basada en Árboles de Decisión (Reglas de producción)
# Se utilizan operadores lógicos y comparaciones múltiples para evaluar el escenario

if mineral == "Oro":
    # El oro suele estar a mayor profundidad y conlleva más riesgo, pero más ganancia
    if profundidad > 500 and nivel_gas < 20:
        # Combinación de alta profundidad con seguridad de gases
        print("\n[PREDICCIÓN IA]: ÉXITO ALTO.")
        print("La veta de oro es profunda pero estable. Se recomienda proceder con la extracción.")
    elif profundidad <= 500 or nivel_gas >= 20:
        # El uso de 'or' permite evaluar si falla alguna de las condiciones de seguridad
        if nivel_gas > 50:
            print("\n[PREDICCIÓN IA]: PELIGRO EXTREMO.")
            print("Niveles de gas tóxico críticos para un minero. Evacuar inmediatamente.")
        else:
            print("\n[PREDICCIÓN IA]: RIESGO MODERADO.")
            print("Profundidad insuficiente para vetas ricas o presencia de gases leves.")
    else:
        print("\n[PREDICCIÓN IA]: DATOS INCONCLUSOS.")

elif mineral == "Carbón":
    # El carbón tiene reglas distintas por su alta volatilidad
    if 100 < profundidad <= 300:
        # Python permite comparaciones múltiples de este tipo para rangos
        if nivel_gas < 10:
            print("\n[PREDICCIÓN IA]: EXTRACCIÓN SEGURA.")
            print("Condiciones óptimas para la minería de carbón en este rango de profundidad.")
        else:
            print("\n[PREDICCIÓN IA]: RIESGO DE EXPLOSIÓN.")
            print("El carbón a esta profundidad con gas detectado es altamente inestable.")
    else:
        print("\n[PREDICCIÓN IA]: BAJA RENTABILIDAD.")
        print("La profundidad no corresponde a los niveles típicos de este mineral.")

elif mineral == "Hierro":
    # Regla simple para minerales de superficie
    if profundidad < 100 and nivel_gas < 5:
        print("\n[PREDICCIÓN IA]: EXTRACCIÓN SIMPLE.")
        print("Operación de superficie segura y de bajo costo.")
    else:
        print("\n[PREDICCIÓN IA]: REVISAR EQUIPAMIENTO.")
        print("Las condiciones no son ideales para una mina de hierro estándar.")

else:
    # Caso para cuando el mineral ingresado no está en la base de conocimientos de la IA
    print("\n[ERROR IA]: MINERAL NO IDENTIFICADO.")
    print("El sistema experto no tiene reglas para el mineral especificado.")
```
